```python
pip install Scrapy
from scrapy import Selector
import requests
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
topstar=[]
result=0
for i in range(10):
 url='https://www.imdb.com/list/ls058011111/?sort=list_order,asc&mode=detail&page={}'.format(i+1)
 html=requests.get(url).content
 sel=Selector(text=html)
 topstar.extend(sel.xpath('//div/h3[@class="lister-item-header"]/a/text()').extract())
result += i
print(result)
```

```python
topstar
```

```python
df=DataFrame(topstar,columns=['Name'])
df=df.replace(r'\n','',regex=True)
df.to_csv('topstar.csv',index=False)
```
